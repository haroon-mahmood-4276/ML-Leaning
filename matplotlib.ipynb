```python
cities = ['Lahore', 'Karachi', 'Multan', 'Islamabad', 'Rawalpindi']
temp = [34, 43, 34, 24, 55]
```

```python
plt.subplot(1,2,1)
plt.xlabel('cities')
plt.ylabel('temp')
plt.title('Hello World')
plt.plot(cities, temp, 'g*')
plt.subplot(1,2,2)
plt.xlabel('cities')
plt.ylabel('temp')
plt.title('Hello World')
plt.plot(x, y, 'r')
```
